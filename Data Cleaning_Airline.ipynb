```python
# ============================================================
# STEP 1: Dim_Date
# ============================================================
dim_date = Delayed_Flights[['Date','year','month','day','DayOfWeek']]\    
           .drop_duplicates()\
           .copy()\
           .reset_index(drop=True)

dim_date.insert(0, 'DateKey', range(1, len(dim_date)+1))    #إدراج عمود جديد 'DateKey' في بداية الجدول 'dim_date' يحتوي على أرقام تسلسلية تبدأ من 1 حتى عدد الصفوف في الجدول، مما يوفر مفتاحًا فريدًا لكل تاريخ في الجدول

print("Dim_Date:", dim_date.shape)   #عرض عدد الصفوف والأعمدة في جدول 'dim_date' بعد إنشاء المفتاح الفريد وإعادة ترتيب الأعمدة
dim_date.head(3)
```

```python
# ============================================================
# STEP 2: Dim_Carrier
# ============================================================
dim_carrier = Delayed_Flights[['UniqueCarrier']]\
              .drop_duplicates()\
              .copy()\
              .reset_index(drop=True)     #تصفية البيانات لاستخراج الأعمدة المتعلقة بشركات الطيران، ثم حذف الصفوف المكررة للحصول على قائمة فريدة من شركات الطيران، ونسخ البيانات إلى إطار بيانات جديد، وإعادة تعيين الفهرس بدون تضمين الفهرس القديم 

dim_carrier.insert(0, 'CarrierKey', range(1, len(dim_carrier)+1))   #إدراج عمود جديد 'CarrierKey' في بداية الجدول 'dim_carrier' يحتوي على أرقام تسلسلية تبدأ من 1 حتى عدد الصفوف في الجدول، مما يوفر مفتاحًا فريدًا لكل شركة طيران في الجدول

print("Dim_Carrier:", dim_carrier.shape)
dim_carrier.head(3)
```

```python
# ============================================================
# STEP 3: Dim_Airport  (Origin + Dest في جدول واحد)
# ============================================================
origins = Delayed_Flights[['Origin']].rename(columns={'Origin': 'AirportCode'})
dests   = Delayed_Flights[['Dest']].rename(columns={'Dest':   'AirportCode'})

dim_airport = pd.concat([origins, dests])\
              .drop_duplicates()\
              .copy()\
              .reset_index(drop=True)

dim_airport.insert(0, 'AirportKey', range(1, len(dim_airport)+1))

print("Dim_Airport:", dim_airport.shape)
dim_airport.head(3)
```

```python
# ============================================================
# STEP 4: Fact_Flights (ربط الـ FK)
# ============================================================
fact = Delayed_Flights.copy()

# ربط DateKey
fact = fact.merge(dim_date[['DateKey','Date','year','month','day','DayOfWeek']],
                  on=['Date','year','month','day','DayOfWeek'], how='left')

# ربط CarrierKey
fact = fact.merge(dim_carrier, on='UniqueCarrier', how='left')

# ربط OriginKey
fact = fact.merge(
    dim_airport.rename(columns={'AirportKey':'OriginKey', 'AirportCode':'Origin'}),
    on='Origin', how='left'
)

# ربط DestKey
fact = fact.merge(
    dim_airport.rename(columns={'AirportKey':'DestKey', 'AirportCode':'Dest'}),
    on='Dest', how='left'
)

# اختار الأعمدة النهائية للـ Fact فقط
fact_flights = fact[[
    'DateKey', 'CarrierKey', 'OriginKey', 'DestKey',
    'FlightNum', 'TailNum',
    'Departure_Delay', 'Arrival_Delay', 'Distance',
    'CarrierDelay', 'WeatherDelay', 'NASDelay',
    'SecurityDelay', 'LateAircraftDelay', 'Delay_sum',
    'ActualElapsedTime', 'Expected_ElapsedTime',
    'AirTime', 'TaxiIn', 'TaxiOut'
]].reset_index(drop=True)

print("Fact_Flights:", fact_flights.shape)
fact_flights.head(3)
```

```python
# ============================================================
# STEP 5: Export للـ CSV
# ============================================================
import os

output_path = r"D:\ITI\power bi qabany\day 6"  # نفس مجلد الـ notebook

dim_date.to_csv(os.path.join(output_path, 'Dim_Date.csv'),    index=False)
dim_carrier.to_csv(os.path.join(output_path, 'Dim_Carrier.csv'), index=False)
dim_airport.to_csv(os.path.join(output_path, 'Dim_Airport.csv'), index=False)
fact_flights.to_csv(os.path.join(output_path, 'Fact_Flights.csv'), index=False)

print("✅ الملفات اتحفظت!")
print(f"  Fact_Flights : {fact_flights.shape[0]:,} صف  |  {fact_flights.shape[1]} عمود")
print(f"  Dim_Date     : {len(dim_date)} تاريخ فريد")
print(f"  Dim_Carrier  : {len(dim_carrier)} شركة طيران")
print(f"  Dim_Airport  : {len(dim_airport)} مطار")
```

```python
import pandas as pd

# تحميل ملف أسماء المطارات من الإنترنت مباشرة
airports_ref = pd.read_csv(
    "https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat",
    header=None,
    names=['AirportID','AirportName','City','Country','IATA','ICAO',
           'Lat','Lon','Alt','Timezone','DST','TzDB','Type','Source'],
    usecols=['AirportName','City','Country','IATA']
)

# فلتر المطارات الأمريكية بس وشيل الـ IATA الفاضية
airports_ref = airports_ref[
    (airports_ref['Country'] == 'United States') &
    (airports_ref['IATA'].notna()) &
    (airports_ref['IATA'] != '\\N')
][['IATA','AirportName','City']].drop_duplicates('IATA')

print(f"✅ عدد المطارات المتاحة: {len(airports_ref)}")
airports_ref.head()
```

```python
# دمج الأسماء مع Dim_Airport
dim_airport = dim_airport.merge(
    airports_ref.rename(columns={'IATA': 'AirportCode'}),
    on='AirportCode',
    how='left'
)

# تحقق من النواقص
missing = dim_airport[dim_airport['AirportName'].isna()]
print(f"⚠️ مطارات مش موجودة أسماؤها: {len(missing)}")
print(missing['AirportCode'].values)

dim_airport.head()
```
